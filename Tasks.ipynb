# `Task 01`
**Write a python program to perform Basic Operations 
load, display and save the images with skimage 
librar**lp:

```python
# First we import io module from skimage and pyplot module from matplotlib
import skimage.io as ski
import matplotlib.pyplot as plt
```

### Load Image

```python
image = ski.imread("Black 3D Artificial Intelligence Youtube Thumbnail.png")
```

### Display

```python
ski.imshow(image)
plt.title("Image")
plt.axis('off')
plt.show()
```

### Save the image

```python
ski.imsave(fname='image.png',arr=image)
```

# `Task 02`
**Write a python program to perform Transformations 
like resize(), rotate() or rescale() on the Images with 
skimage librar. **

```python
## For transformation we need to load transform module from skimage
import skimage.transform as tfm
```

### Resize

```python
# First we check the original size of image
o_shape = image.shape
o_shape
```

```python
## Now give it a resized shape
r_shape = (o_shape[0]//3, o_shape[1]//3, o_shape[2]) # We don't change channel
r_image = tfm.resize(image, r_shape)
```

```python
r_image.shape
```

```python
ski.imshow(r_image)
plt.title("Resized Image")
plt.axis('off')
plt.show()
```

**Now it's look like blured b/c we reduze its size 3 time**

### Rotate

```python
i_m_c = tfm.rotate(image, -90) 
i_m_a = tfm.rotate(image, 90)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, figsize=(8, 4),sharex=True,sharey=True)
fig.suptitle("Rotated Image")
fig.tight_layout()

ax1.imshow(image)
ax1.set_title("Original Image")
ax1.axis('off')

ax2.imshow(i_m_c)
ax2.set_title("Clock Wise Rotated Image")
ax2.axis('off')

ax3.imshow(i_m_a)
ax3.set_title("Anti Clock Wise Rotated Image")
ax3.axis('off')

plt.show()
```

### Rescale

```python
image_rescaled = tfm.rescale(image, 1/4,anti_aliasing=True)
ski.imshow(image_rescaled)
plt.title('Original Image Rescaled by One-Fourth\n with Anti-Aliasing')
plt.axis('off')
plt.show()
```

# `Task 03`
**Python script to ignore low intensity pixels in 
an imag** 

```python
# import libraries to ignore low imn
import imageio.v3 as iio
```

```python
image = iio.imread('image.png')
```

```python
ski.imshow(image)
plt.title("Image Loaded Using imageio.v3")
plt.axis('off')
plt.show()
```

```python
image.shape
```

```python
import numpy as np
```

```python
maize_roots = np.array(image)
```

```python
maize_roots[0]
```

```python
# display original image
fig, ax = plt.subplots()
ax.imshow(maize_roots)
plt.axis('off')
plt.show()
```

```python
# keep only high-intensity pixels
maize_roots[maize_roots < 128] = 0
```

```python
# display modified image
fig, ax = plt.subplots()
ax.imshow(maize_roots)
plt.title("Modified Image")
plt.axis('off')
plt.show()
```
