### Import Libraries

```python
import skimage.io
import matplotlib.pyplot as plt
```

```python
image = skimage.io.imread("image.jfif")
```

## Load Image as a GrayScale

```python
image = skimage.io.imread("image.jfif",as_gray=True)
skimage.io.imshow(image)
plt.title("Gray Scale Image")
plt.axis('off')
plt.show()
```

## Load Color Image and Then Turn Into Gray

```python
import skimage.io as ski
import skimage as sk
```

```python
image = ski.imread('image.jfif')
```

```python
# Show Image
ski.imshow(image)
plt.title("ColorFul Image")
plt.axis('off')
plt.show()
```

```python
## Turn it into grayscale
image = sk.color.rgb2gray(image)
```

```python
ski.imshow(image)
plt.title("Gray Scale")
plt.axis('off')
plt.show()
```

```python
# Size of the image
image.shape
```

```python
## Image Height
print("Height ",image.shape[0])
print("Width ",image.shape[1])
```

```python
#We will cut the height and the width of the image by half.
output_shape=(image.shape[0]//2, image.shape[1]//2)
```

```python
resized_image = sk.transform.resize(image, output_shape)
```

```python
resized_image.shape
```

```python
ski.imshow(resized_image)
plt.title("Resized Image")
plt.axis("off")
plt.show()
```

**When We will cut the height and the width of the image by half.**


## Rotate

```python
image_rotated_clockwise = sk.transform.rotate(image, -90) 
image_rotated_anticlockwise = sk.transform.rotate(image, 90)
#Display Original, Clockwise and Anticlockwise rotated images.
fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, figsize=(12, 10),sharex=True,sharey=True)
fig.tight_layout()
ax1.imshow(image)
ax1.set_title('Original Image')
ax1.axis('off')
ax2.imshow(image_rotated_clockwise)
ax2.set_title('Image Rotated \n 90 Degrees Clockwise:')
ax2.axis('off')
ax3.imshow(image_rotated_anticlockwise)
ax3.set_title('Image Rotated \n 90 Degrees Anticlockwise:')
ax3.axis('off')
plt.show()
```

## Rescale

```python
image_rescaled = sk.transform.rescale(image, 1/4,anti_aliasing=False)
ski.imshow(image_rescaled)
plt.title('Original Image Rescaled by One-Fourth\n without Anti-Aliasing')
plt.axis('off')
plt.show()
```
